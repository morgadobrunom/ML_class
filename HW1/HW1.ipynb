# Question 4

```python
from sklearn import datasets

iris = datasets.load_iris()
X = iris.data[:, :3] # only selecting first three features
y = iris.target


from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test=train_test_split(
    X,y,
    test_size=0.40,
    train_size=0.60,
    random_state=123,
    shuffle=True,
    stratify=y)

#Importing the SVM
from sklearn import svm

#Training SVM
clf = svm.SVC()
clf.fit(X_train, y_train)

#Extracting data from prodicted
preds = clf.predict(X_test)

#Evaluate 
from sklearn.metrics import accuracy_score
acc = accuracy_score(y_test,preds)
print('accuracy score :', acc)
```

This accuracy rate is much higher than the two feature version of 0.77 from the in-class example.

# Question 5

```python
import tensorflow as tf
from numpy import loadtxt
from keras.models import Sequential
from keras.layers import Dense
from keras import optimizers
from tensorflow.python.keras.optimizers import *
import numpy as np

#Making input and output data
X = np.arange(1, 101, 1)
X = X*1.0
print("input data array: ")
print(X)

y = X**(1/7)
print()
print("output data array: ")
print(y)
```

```python
#define keras model
model = Sequential()

model.add(Dense(3,input_dim=1,activation='elu'))
model.add(Dense(6,activation='elu'))
model.add(Dense(3,activation='elu'))
model.add(Dense(1))

#compile the keras model
opt = optimizers.Adam(learning_rate=0.001)
mse = tf.keras.losses.MeanSquaredError(
    reduction=tf.keras.losses.Reduction.SUM)
model.compile(loss=mse, optimizer=opt)



#fit the keras model on the dataset (CPU)
model.fit(X,y,epochs=2000,batch_size=10, verbose=0)
model.summary()

#make class predictions with the model
predictions = model.predict(X)

```

```python
#summarize the first every 10 cases
for i in range(10):
    print('%.1f => %.2f (expected %.2f)' %(X[10*i], predictions[10*i], y[10*i]) )
```

```python
import matplotlib.pyplot as plt
number_grid = np.linspace(1, 100, 100)
plt.scatter(X,y, label='data')
plt.plot(number_grid,model.predict(np.expand_dims(number_grid,axis=1)) , color='red', label='model')
plt.xlabel('number')
plt.ylabel('seventh root')
plt.legend()
```

For this data I changed the activation function from ReLu to ELU which has a shape more reminiscent of the data. I also increased the number of data points to 100. This allows me to reduce the number of neurons per layer from 6 to 3 and still have pretty good accuracy.
